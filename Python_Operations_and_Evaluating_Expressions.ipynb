<h3>Mixing Data Types</h3>
<p style="text-align:justify;">
Python has many built-in <a href="https://www.w3schools.com/python/python_datatypes.asp">data types</a> and we need to know what happens when we mix different data types while using arithmetic operators. The rule of thumb while mixing data types is, <u>the larger data type always wins</u>.
</p>
<p>
    Here are some of the data-types of interest from largeset to smallest:
<ol>
    <li>String (largest)</li>
    <li>Float</li>
    <li>Integer (smallest)</li>
</ol>

```python
# Here are some legal examples of mixing data-types with arithmetic operators.

print('1 + 2.0 =', 1 + 2.0)     # Integer + Float => Float
print('3 * "2" =', 3 * "2")     # Integer * String => String


# We can always mix numeric data types, but there are limitations while mixing other data types.
# ^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^
# For exemple, we can't:
# - Add Strings and Numbers
# - Subtract Strings and Numbers
# - Divide Strings and Numbers
```

<h3>Evaluating Expressions</h3>
<p style="text-align:justify;">
While evaluating expressions in code, the typical mathematical heirarchy rules apply (PEMDAS).<br /><br />
However, we have to keep two additional things in mind:
<ol>
    <li>Each operation's respective data types will dictate the resulting data type.</li>
    <li>The interpreter is going to read from left to right, unless otherwise dictated.</li>
</ol>
